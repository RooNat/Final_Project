# Google Trends Processing

Data Source: [Bitcoin Historical prices & Google Trends data](https://www.kaggle.com/datasets/d947f5a4afa2d5f4de04839ffb8f018ff6807a238b80ad77d3baa0fa01158af5?select=Historical+interest+of+bitcoin+%282016-2022%29.csv)


```python
import pandas as pd
import numpy as np
import datetime
from datetime import datetime
```

```python
google_trends= pd.read_csv("./data/Google Trends/Historical interest of bitcoin (2016-2022).csv")
```

```python
# Define the format of the input date-time string
input_format = "%d/%m/%Y %H:%M"
google_trends['date']= google_trends['date'].apply(lambda x: datetime.strptime(x, input_format))
```

```python
df_sorted=google_trends.sort_values(by='date',ignore_index=True)
```

```python
df_sorted
```

```python
# filter the data based on the year
df_2022=df_sorted[df_sorted['date'].dt.year==2022]
df_2022_1=df_2022[df_2022['date'].dt.month==1]
df_2022_1
```

```python
df_2022.to_csv("./data/Google Trends/Trends_2022.csv",index=False)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(df_2022_1['date'], df_2022_1['Value'])
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Google Trends')
plt.grid(True)
plt.tight_layout()
plt.show()
```
