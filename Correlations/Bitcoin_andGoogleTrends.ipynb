# Google Trends Selection

```python
import pandas as pd
import numpy as np
```

```python
close_data=pd.read_csv('./../data/processed_data.csv')
close_data.rename(columns={'Unnamed: 0': 'datetime'},inplace=True)
close_data['datetime']=pd.to_datetime(close_data['datetime'])
close_data.set_index('datetime',inplace=True)
display(close_data)
```

```python
google_data=pd.read_csv('./../data/Google Trends/Trends_replace.csv')
google_data_non=pd.read_csv('./../data/Google Trends/Trend_non_replace.csv')
```

```python
google_data['date']= pd.to_datetime(google_data['date'])
google_data_non['date']= pd.to_datetime(google_data_non['date'])
google_data.set_index('date',inplace=True)
google_data_non.set_index('date',inplace=True)
```

```python
display(google_data)
```

```python
def get_data(google_data, google_data_non, close_data, month):
    sub_data_1=google_data[(google_data.index.month>=1) & (google_data.index.month <=month)]
    sub_data_2=google_data_non[(google_data_non.index.month>=1) & (google_data_non.index.month <=month)]
    sub_close_data=close_data[(close_data.index.month>=1) & (close_data.index.month <=month)]
    sub_data_1.rename(columns={'Scaled Value':'Google_Trends'}, inplace=True)
    sub_data_2.rename(columns={'Scaled Value':'Google_Trends_non'}, inplace=True)
    bitcoindata=pd.concat([sub_close_data,sub_data_1,sub_data_2], axis=1)
    # keep the useful columns
    columns_keep=['Close','Google_Trends','Google_Trends_non','Bitcoin']
    columns=list(bitcoindata.columns.values)
    print(columns)
    for i in columns_keep:
        columns.remove(i)

    bitcoin_new=bitcoindata.drop(columns=columns)
    bitcoin_new_google=bitcoin_new.drop(columns=['Bitcoin'])
    return bitcoin_new_google, bitcoin_new
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
import plotly.express as px
# sns.heatmap(bitcoin_new_google.corr(method='spearman'), annot=True)
month=[1,2,3,4,5,6]
from tqdm import tqdm
results=[]
for m in tqdm(month, total=len(month)):
    data, bitcoin_new=get_data(google_data, google_data_non,close_data, m)
    correlation=data.corr(method='spearman')
    results.append({"Months":m, "Google_trends_replace": correlation['Close']['Google_Trends'],"Google_trends_non_replace": correlation['Close']['Google_Trends_non']})
```

```python
results
```

```python
results_frame=pd.DataFrame(results)
```

```python
results_frame.set_index('Months',inplace=True)
results_frame
```

```python
# import plotly.graph_objects as go
# fig = go.Figure()

fig = px.bar(results_frame, x=results_frame.index, y=['Google_trends_replace', 'Google_trends_non_replace'], title='Google Trends')

fig.update_layout(
    title='The correlation between Google Trends and Close Price',
    xaxis_title='Months',
    yaxis_title='Spearman Correlation',
    barmode='group'
)

fig.show()
```

```python
month=1
sub_data_1=google_data[(google_data.index.month>=month) & (google_data.index.month <=month)]
sub_data_2=google_data_non[(google_data_non.index.month>=month) & (google_data_non.index.month <=month)]
sub_close_data=close_data[(close_data.index.month>=month) & (close_data.index.month <=month)]
sub_data_1.rename(columns={'Scaled Value':'Google_Trends'}, inplace=True)
sub_data_2.rename(columns={'Scaled Value':'Google_Trends_non'}, inplace=True)
bitcoindata=pd.concat([sub_close_data,sub_data_1,sub_data_2], axis=1)
# keep the useful columns
columns_keep=['Close','Google_Trends','Google_Trends_non','Bitcoin']
columns=list(bitcoindata.columns.values)
print(columns)
for i in columns_keep:
    columns.remove(i)

bitcoin_new=bitcoindata.drop(columns=columns)
bitcoin_new_google=bitcoin_new.drop(columns=['Bitcoin'])
```

```python
bitcoin_google_trans=bitcoin_new.drop(columns=['Close','Google_Trends_non'])
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(bitcoin_google_trans)
scaled_df = pd.DataFrame(scaled_data, columns=['Adjusted Google Trends','Original Google Trends'],index=bitcoin_google_trans.index)
```

```python
fig = px.line(scaled_df, x=scaled_df.index, y=scaled_df.columns, title='Bitcoin Google Trends')
fig.update_layout(legend_title_text='Variables',width=800, height=600)
fig.update_xaxes(title_text="Datetime")
# Set y-axes titles
fig.update_yaxes(title_text="Google Trends", secondary_y=False)
fig.show()
```

```python
# Only show Jan
scaled_df_1=scaled_df[scaled_df.index.month==1]
fig = px.line(scaled_df_1, x=scaled_df_1.index, y=scaled_df_1.columns, title='Bitcoin Google Trends - 1 month')
fig.update_layout(legend_title_text='Variables',width=1000, height=800)
fig.update_xaxes(title_text="Datetime")
# Set y-axes titles
fig.update_yaxes(title_text="Google Trends", secondary_y=False)
fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from itertools import cycle
def plot_chart(x,y1,y2,name1,name2,axes1,axes2):
    # plot the data of the google trends and Bitcoin
    fig = make_subplots(specs=[[{"secondary_y": True}]])
    fig.add_trace(go.Scatter(x=x,
                             y=y1,
                             mode='lines',
                             name=name1), secondary_y=False)
    fig.add_trace(go.Scatter(x=x,
                             y=y2,
                             mode='lines',
                             name=name2), secondary_y=True)
    # Add axis titles
    fig.update_layout(
        title_text="Bitcoin and Google Trends",
        width=1000,
        height=600,
    )

    fig.update_xaxes(title_text="Datetime")

    # Set y-axes titles
    fig.update_yaxes(title_text=axes1, secondary_y=False)
    fig.update_yaxes(title_text=axes2, secondary_y=True)
    fig.show()
```

```python
price_google=bitcoin_new_google.drop(columns=['Google_Trends_non'])
price_google
```

```python
x= price_google.index
y1=price_google['Close']
y2=price_google['Google_Trends']
name1="Bitcoin Price"
name2="Google Trends"
axes1= "Bitcoin Price"
axes2="Google Trends"
plot_chart(x, y1, y2, name1, name2, axes1, axes2)
```
